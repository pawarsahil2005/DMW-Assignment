### Removing Outliers

```python
Q1 = df1['Age'].quantile(0.25)
Q3 = df1['Age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

### Central Tendency & Dispersion

```python
numeric_df = df3.select_dtypes(include=['number'])

# Central Tendency 
mean_values = numeric_df.mean()
median_values = numeric_df.median()
mode_values = numeric_df.mode().iloc[0]

print("\nMean:\n", mean_values)
print("\nMedian:\n", median_values)
print("\nMode:\n", mode_values)
```

```python
# Dispersion 
std_values = numeric_df.std()
var_values = numeric_df.var()
range_values = numeric_df.max() - numeric_df.min()

print("\nStandard Deviation:\n", std_values)
print("\nVariance:\n", var_values)
print("\nRange:\n", range_values)
```
